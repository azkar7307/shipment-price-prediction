```python
def seperate_numerical_categorical_columns(df):
        numerical_columns = []
        categorical_columns = []

        for column in df.columns:
            if df[column].dtypes == 'int64' or df[column] == 'float64':
                numerical_columns.append(column)
            
            else:
                categorical_columns.append(column)
        
        return numerical_columns, categorical_columns
```

```python
seperate_numerical_categorical_columns(df)
```

```python
def seperate_numerical_categorical_columns(df):
    numerical_columns = df.select_dtypes(include=np.number).columns.tolist()
    categorical_columns = df.select_dtypes(include=[object, pd.Categorical]).columns.tolist()

        
    return numerical_columns, categorical_columns
```
